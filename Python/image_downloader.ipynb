```python
import os
import requests

# Function to download and save an image from a URL to the desktop
def download_image_from_url(url, image_name):
    try:
        # Send a GET request to the URL to fetch the image content
        response = requests.get(url)
        
        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            # Determine the desktop directory
            desktop_dir = os.path.join(os.path.expanduser('~'), 'Desktop')
            
            # Create the directory if it doesn't exist
            if not os.path.exists(desktop_dir):
                os.makedirs(desktop_dir)
            
            # Create the full path for saving the image
            image_path = os.path.join(desktop_dir, image_name)
            
            # Save the image to the desktop
            with open(image_path, 'wb') as image_file:
                image_file.write(response.content)
            
            print(f"Image '{image_name}' has been downloaded and saved to the desktop.")
        else:
            print(f"Failed to download image. Status code: {response.status_code}")
    except Exception as e:
        print(f"An error occurred: {str(e)}")

# Example usage:
if __name__ == "__main__":
    image_url = "https://dfstudio-d420.kxcdn.com/wordpress/wp-content/uploads/2019/06/digital_camera_photo-1080x675.jpg"  # Replace with the URL of the image you want to download
    image_name = "downloaded_image.jpg"  # Replace with the desired name for the saved image
    download_image_from_url(image_url, image_name)
```
